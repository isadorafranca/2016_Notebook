```r
library("ggplot2")
library("Hmisc")
setwd("~/Dropbox (Silliman)/LinuxBackup/Projects/Alpheus/Fst/")
```

```r
fst = read.csv("Alpheus GPhocs and Stats - FST.csv",header=TRUE)
A2c4sp2 <- fst[which(fst$Dataset == "A2c4nsp2"), c("Pair","ML.phylo.distance","mean_FST","X..nucs.shared","X..fixed")]
A2c4sp2 <- A2c4sp2[which(is.na(A2c4sp2$ML.phylo.distance) != TRUE),]
```

```r
HP = ggplot(A2c4sp2, aes(x = X..nucs.shared, y = ML.phylo.distance))
HP + geom_point() + xlab(expression(atop("Coverage",atop("# of bases sequenced in both species")))) + ylab("ML Phylogenetic Distance") +ggtitle(expression(atop("Pairwise Phylogenetic Distance and # of Shared Nucleotides", atop(italic("A2c4nsp2 Dataset", "")))))
```

```r
rspear <- rcorr(as.matrix(A2c4sp2[,c("X..nucs.shared","ML.phylo.distance")]), type="spearman")
rspear[[1]]
rspear$P
```

```r
pspear <- rcorr(as.matrix(A2c4sp2[,c("X..nucs.shared","ML.phylo.distance")]), type="pearson")
pspear[[1]]
pspear$P
```

```r
HP = ggplot(A2c4sp2, aes(x = mean_FST, y = ML.phylo.distance))
HP + geom_point() + xlab("Mean FST") + ylab("ML Phylogenetic Distance") +ggtitle(expression(atop("Pairwise Phylogenetic Distance and Mean FST", atop(italic("A2c4nsp2 Dataset", "")))))
```
